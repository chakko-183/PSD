# **forecasting**

## Apa Itu Forecasting?

Forecasting (Peramalan) adalah proses memprediksi nilai atau peristiwa di masa depan berdasarkan data masa lalu dan/atau faktor saat ini.

Intinya, forecasting menggunakan pola historis, tren, dan hubungan antar variabel untuk membuat perkiraan yang terinformasi tentang apa yang mungkin terjadi di waktu mendatang.

Dalam konteks matematika dan data sains, forecasting sering dikaitkan dengan Analisis Deret Waktu (Time Series Analysis), di mana data diurutkan secara kronologis (misalnya, harga saham harian, penjualan bulanan, atau suhu tahunan).

## Mengapa Forecasting Itu Penting?

Forecasting sangat penting karena memungkinkan pengambilan keputusan yang proaktif dan alokasi sumber daya yang efisien. Tanpa peramalan, organisasi hanya dapat bereaksi terhadap kejadian, yang dapat menyebabkan pemborosan atau kehilangan peluang.

## Jenis-Jenis Forecasting
Metode forecasting secara umum dapat dibagi menjadi dua kategori besar berdasarkan pendekatan datanya: Kualitatif dan Kuantitatif.

1.   Forecasting Kuantitatif

    Metode ini bergantung pada data historis dan model matematis. Ini cocok ketika tersedia banyak data masa lalu yang relevan.

2.   Forecasting Kualitatif

    Metode ini digunakan ketika data historis tidak tersedia, tidak relevan, atau ketika situasi sangat unik sehingga dibutuhkan penilaian ahli atau intuisi.
