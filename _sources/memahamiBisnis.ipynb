# **Memahami Bisnis**

Business Understanding adalah tahap awal proyek data/ML di mana tim memahami konteks bisnis, tujuan, pemangku kepentingan, batasan, dan bagaimana keberhasilan akan diukur. Ini menjembatani masalah bisnis nyata dengan kebutuhan teknis analisis data.

1. Tujuan Bisnis (contoh untuk Superstore)

  * Meningkatkan profitabilitas — identifikasi produk/segmen/region yang   memberi margin terbesar/terkecil.

  * Mengurangi pengiriman terlambat — identifikasi penyebab ship delays dan optimalkan logistik.

  * Meningkatkan retensi pelanggan — temukan pola pelanggan bernilai tinggi.

  * Optimasi stok & kategori produk — kurangi stok mati, tingkatkan bestseller.
2. Pertanyaan Bisnis Kunci (business questions)

  * Produk atau sub-kategori mana yang paling sering rugi?

  * Region mana yang memiliki rata-rata profit per order terendah?

3. Pemangku Kepentingan (Stakeholders)

  * Product Manager / Head of Sales — pengambilan keputusan produk & promosi.

  * Logistik / Operasional — pengiriman dan lead time.

  * Finance — margin, profit/loss.

  * Marketing — segmentasi pelanggan & kampanye.
4. Kriteria Keberhasilan (KPIs & Success Metrics)

  Contoh KPI terukur:

    * Total Profit (per bulan / quarter) — naik X% dalam 6 bulan.

    * Profit Margin per kategori — target M% minimal.

    * On-time Shipment Rate — target ≥ 95%.

    * Customer Lifetime Value (CLTV) peningkatan.
  5. Batasan & Asumsi

  * Kualitas data: asumsi bahwa data transaksi lengkap untuk periode analisis.

  * Konektivitas & privasi: data sensitif (customer name/ID) tidak dibagikan sembarangan.

  * Waktu: analisis hanya untuk rentang tanggal tertentu (mis. 2017-2018).
6. Output Bisnis yang Diharapkan

  * Dashboard KPI (sales, profit by region/category).

  * Rekomendasi pricing/discount policy berdasarkan efektivitas diskon.

  * Daftar aksi prioritas (mis. hapus item loss-making, optimasi shipping).
